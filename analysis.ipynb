```python
df = pd.read_csv(r"C:\Users\Hp\Downloads\archive (2)\uber_trips_dataset_50k.csv")

df['pickup_time'] = pd.to_datetime(df['pickup_time'], errors='coerce')
df = df.dropna(subset=['pickup_time'])

df['hour'] = df['pickup_time'].dt.hour
df['weekday'] = df['pickup_time'].dt.day_name()

df = df[df['status'] == 'Completed']
df = df[df['fare_amount'] > 0]

print(df.shape)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,5))
sns.countplot(x='hour', data=df)
plt.title("Peak Booking Hours")
plt.xlabel("Hour")
plt.ylabel("Number of Rides")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['pickup_lng'], df['pickup_lat'], alpha=0.3)
plt.title("Pickup Locations (Demand Zones)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

```python
sns.kdeplot(
    x=df['pickup_lng'],
    y=df['pickup_lat'],
    fill=True,
    cmap="Reds"
)
plt.title("High Demand Areas (Heatmap)")
plt.show()
```

```python
sns.scatterplot(x='distance_km', y='fare_amount', data=df)
plt.title("Distance vs Fare")
plt.show()
```

```python
sns.regplot(x='distance_km', y='fare_amount', data=df, scatter_kws={'alpha':0.3})
plt.title("Distance vs Fare with Trend Line")
plt.show()
```

```python
avg_fare = df.groupby('hour')['fare_amount'].mean()

avg_fare.plot(kind='line', figsize=(10,5))
plt.title("Average Fare by Hour")
plt.xlabel("Hour")
plt.ylabel("Average Fare")
plt.show()
```
